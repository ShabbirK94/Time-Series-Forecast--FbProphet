# #4: PREPARING THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado_df
```

```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']]
```

```python
avocado_prophet_df 
```

```python
#But facebook prophet only takes values in terms of 'ds' and 'y'
avocado_prophet_df = avocado_prophet_df.rename(columns= {'Date': 'ds', 'AveragePrice': 'y'})
avocado_prophet_df 
```

# #5: UNDERSTANDING THE INTUITION BEHIND FACEBOOK PROPHET

# #6: DEVELOPING THE MODEL AND MAKING PREDICTIONS - PART A

```python
m = Prophet()
m.fit(avocado_prophet_df)
```

```python
# Forcasting into the future
future = m.make_future_dataframe(periods= 365)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast, xlabel= 'Date', ylabel = 'Price')
```

```python
figure2 =m.plot_components(forecast)
```

# #7: DEVELOPING MODEL AND MAKING PREDICTIONS (REGION: CHICAGO) - PART B

```python
# dataframes creation for both training and testing datasets 
avocado_df = pd.read_csv('avocado.csv')
```

```python
# Select specific region
avocado_df_chicago= avocado_df[avocado_df['region']=='Chicago']
```

```python
avacado_df_chicago= avocado_df_chicago.sort_values('Date')
```

```python
plt.plot(avocado_df_chicago['Date'], avocado_df_chicago['AveragePrice'])
```

```python
avocado_df_chicago= avocado_df_chicago.rename(columns={'Date':'ds', 'AveragePrice': 'y'})
```

```python
m = Prophet()
m.fit(avocado_df_chicago)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```

So, we see that the avocado price trends projected in the region of Chicago is quite different from the overall trend.
